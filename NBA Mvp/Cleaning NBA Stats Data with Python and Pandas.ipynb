```python
def single_team(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Tm"]=="TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row

    
players = players.groupby(["Player","Year"]).apply(single_team)
```

```python
players.head()
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
combined = players.merge(mvps, how = "outer", on = ["Player", "Year"])
```

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
teams = pd.read_csv(r"C:\Users\Windows10\OneDrive - ust.edu.ph\Ust Engineering\Titles & Certifications\Certifications\Google Data Analytics\Data Analysis Projects\NBA Mvp\teams.csv")
```

```python
teams.head()
```

```python
teams = teams[~teams["W"].str.contains("Division")].copy()
```

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex = False)
```

```python
sorted(teams["Team"].unique())
```

```python
sorted(combined["Tm"].unique())
```

```python
nicknames = {}
with open(r"C:\Users\Windows10\OneDrive - ust.edu.ph\Ust Engineering\Titles & Certifications\Certifications\Google Data Analytics\Data Analysis Projects\NBA Mvp\nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev,name = line.replace("\n","").split(",")
        nicknames[abbrev] = name
```

```python
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
combined.head()
```

```python
train = combined.merge(teams, how="outer", on=["Team", "Year"])
```
